```python
from scipy.linalg import hadamard

def hadamard_transform(img):
    if img.shape[0] != img.shape[1]:
        raise ValueError('Image shape must be squared')
    had = hadamard(img.shape[0])
    return np.dot(img, had) / img.shape[0]
```
